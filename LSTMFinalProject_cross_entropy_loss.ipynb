## Categorical variables processing

```python
categorical_cols = df.select_dtypes(include='object').columns

print(f'Categorical columns are {categorical_cols.values}')
```

The weather usually depends on lot of factors including **seasonality**. We need to properly handly date field to capture this.

1. One way is to convert to unix format (ignores seasonality but faster processing)
2. Cyclical Encoding. In this we take sin(day), sin(month) and cos(day), cos(month).

What this essentially means is that, if we have 2 dates day 1 of next month and day 31 current month. The models learns that there are 30 day difference if cyclic incode is not given. If cyclic encoding is used, it essentiall gives a similar value for both of them (means closer).

Below is an representation of day and month being cyclic in nature (for our dataset)

```python
df['Date'] = pd.to_datetime(df['Date'])

ax = df['Date'].dt.day[:365].plot()
ax.set_title("Days distribution for an year")
ax.set_xlabel("days in an year")
ax.set_ylabel("days in a month")
```

```python
df['Year'] = df['Date'].dt.year

df['sin_month'] = np.sin(2*np.pi*df['Date'].dt.month/12)
df['cos_month'] = np.cos(2*np.pi*df['Date'].dt.month/12)

df['sin_day'] = np.sin(2*np.pi*df['Date'].dt.day/31)
df['cos_day'] = np.cos(2*np.pi*df['Date'].dt.day/31)

# We should drop date colum as it is unnecessary (dropped later)
```

```python
df[categorical_cols].describe()
```

```python
(df[categorical_cols].isna().mean()*100).sort_values(ascending=False)
```

We have a maz of 7.26% null values in categories. We can use mode to fill them.

```python
for col in df[categorical_cols].columns:
  df[col] = df[col].ffill()

# Checking nulls after imputation
(df[categorical_cols].isna().mean()*100).sort_values(ascending=False)
```

From above table, we can understand that Date is not unique throughout column but to each location. Weather is location specific. It essentially means we have 49 time series problems. There are 2 ways 2 solve this,
1. using 49 LSTM models
2. use one LSTM model but encode location (one-hot encoding - 49 new features)

The second method is more efficient if we efficiently handly the one-hot encoding part. So instead of doing onehot encoding, we can do embedding (converting 49 columns to 10 columns(say)).

For Location feature - 2 steps are followed:
1. label encoding (49 unique values)
2. Embedding layer while building lstm model

For other categories, we can use label encoder.

```python
for col in df[categorical_cols].columns:
  if col != 'Date':
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
# All columns are now converted numerically (except Location column - handled later)
df[categorical_cols].describe()
```

```python
n_features = df_scaled.shape[1]
n_features
```

```python
print(f'Shape of features is {features.shape} and shape of target is {target.shape}')
```

Create LSTM Sequence

```python
# LSTM input format: (n_samples, timesteps, features)
# For our case: (n_samples, 7, 24)  # Timesteps is lookback - We are using 7 days of data to predict rainfall next day

# predictive horizon = 1 by default (predicts next day rain)
def create_sequences(data, timesteps, target, ph=1):
    X, y = [], []
    for i in range(len(data) - timesteps - ph + 1):  # Adjust range for horizon
        X.append(data[i:i + timesteps])  # Shape: (timesteps, 24)
        y.append(target[i + timesteps + ph - 1])  # Target shifted by ph
    return np.array(X), np.array(y)

X, y = create_sequences(features, timesteps, target, ph)
```

```python
print(f'Shape of X is {X.shape} and shape of y is {y.shape}')
```

```python
X_scaled = X[:, :, :23]  # Scaled features
X_location = X[:, :, 23]  # Location
```

```python
input_scaled = Input(shape=(timesteps, n_features - 1)) # removing 1 feature (Location)
input_location = Input(shape=(timesteps,))

location_embedding = Embedding(input_dim=len(np.unique(X_location)), output_dim=24)(input_location) # 49 col is converted to 20 cols
combined_input = Concatenate(axis=-1)([input_scaled, location_embedding])

lstm_1 = Bidirectional(LSTM(32, return_sequences=False, kernel_regularizer=l2(0.01)))(combined_input)
dropout_1 = Dropout(0.4)(lstm_1)
# lstm_2 = Bidirectional(LSTM(32, return_sequences=False, kernel_regularizer=l2(0.01)))(dropout_1)
# dropout_2 = Dropout(0.4)(lstm_2)
dense = Dense(16, activation='relu', kernel_regularizer=l2(0.01))(dropout_1)
output = Dense(1, activation='sigmoid')(dense)
model = Model(inputs=[input_scaled, input_location], outputs=output)

optimizer = Adam(learning_rate=LEARNING_RATE)
# Compile
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
total_samples = len(X_scaled)
train_split = 0.7
val_split = 0.15
# Test split is the remainder (0.15)

train_size = int(train_split * total_samples)
val_size = int(val_split * total_samples)
test_size = total_samples - train_size - val_size

#Sequential split (no shuffling)
X_train_scaled = X_scaled[:train_size]
X_val_scaled = X_scaled[train_size:train_size + val_size]
X_test_scaled = X_scaled[train_size + val_size:]

X_train_loc = X_location[:train_size]
X_val_loc = X_location[train_size:train_size + val_size]
X_test_loc = X_location[train_size + val_size:]

y_train = y[:train_size]
y_val = y[train_size:train_size + val_size]
y_test = y[train_size + val_size:]

#Create tf.data.Datasets
train_dataset = tf.data.Dataset.from_tensor_slices(((X_train_scaled, X_train_loc), y_train))
val_dataset = tf.data.Dataset.from_tensor_slices(((X_val_scaled, X_val_loc), y_val))
test_dataset = tf.data.Dataset.from_tensor_slices(((X_test_scaled, X_test_loc), y_test))

#Batch and prefetch (no shuffling for sequence data)
train_dataset = train_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_dataset = val_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Verify sizes
print(f"Train samples: {train_size}, Val samples: {val_size}, Test samples: {test_size}")
```

```python
# Compute class weights
class_weights = compute_class_weight(class_weight="balanced", classes=np.unique(y_train), y=y_train)
class_weights
```

```python
class_weight_dict = {0: class_weights[0], 1: class_weights[1]*0.4}
```

```python
# Check if GPU is detected
print("Num GPUs Available:", len(tf.config.list_physical_devices('GPU')))

# Get GPU details
!nvidia-smi
```

```python
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-06)
```

```python
# Train
with tf.device('/GPU:0'):
  history =  model.fit(
      train_dataset,
      epochs=EPOCH,
      batch_size=BATCH_SIZE,
      validation_data=val_dataset,
       class_weight=class_weight_dict,
      callbacks=[tf.keras.callbacks.EarlyStopping(patience=20, restore_best_weights=True),lr_scheduler ]
  )
```

```python
import matplotlib.pyplot as plt

# Plot loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.savefig('val_loss_plot.png', dpi=300, bbox_inches='tight')
plt.legend()

# Plot accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.savefig('val_accuracy_plot.png', dpi=300, bbox_inches='tight')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Predictions
y_pred_prob = model.predict(test_dataset)
y_pred = (y_pred_prob > 0.45).astype(int).flatten()  # Binary predictions

# Extract y_test_flat safely
y_test_flat = np.array([y.numpy() for _, y in test_dataset.unbatch()])  # Convert tensors to numpy
print("y_test_flat shape:", y_test_flat.shape)
```

```python
# Classification report
print("Classification Report:")
print(classification_report(y_test_flat, y_pred, target_names=['No Rain', 'Rain']))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test_flat, y_pred)
print("Confusion Matrix:")
cm
```
